# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

# Load dataset 

```python
datasets.load_iris().keys()
```

```python
datasets.load_iris()['target_names']
```

```python
datasets.load_iris()['feature_names']
```

```python
data = datasets.load_iris()
df = pd.DataFrame(data['data'], columns=data['feature_names'])
df['target'] = data['target']
df['target_name'] = df['target'].map(lambda x: data['target_names'][x])
df.head()
```

# Scatter Plot of the Iris dataset

```python
_, ax = plt.subplots(2, 3, figsize=(16, 8))

ax[0, 0].scatter(df.iloc[:, 0], df.iloc[:, 1], c=df.target)
ax[0, 0].set(xlabel=df.columns[0], ylabel=df.columns[1])
ax[0, 1].scatter(df.iloc[:, 0], df.iloc[:, 2], c=df.target)
ax[0, 1].set(xlabel=df.columns[0], ylabel=df.columns[2])
ax[0, 2].scatter(df.iloc[:, 0], df.iloc[:, 3], c=df.target)
ax[0, 2].set(xlabel=df.columns[0], ylabel=df.columns[3])
ax[1, 0].scatter(df.iloc[:, 1], df.iloc[:, 2], c=df.target)
ax[1, 0].set(xlabel=df.columns[1], ylabel=df.columns[2])
ax[1, 1].scatter(df.iloc[:, 1], df.iloc[:, 3], c=df.target)
ax[1, 1].set(xlabel=df.columns[1], ylabel=df.columns[3])
ax[1, 2].scatter(df.iloc[:, 2], df.iloc[:, 3], c=df.target)
ax[1, 2].set(xlabel=df.columns[2], ylabel=df.columns[3]);
```

# Model

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :4], df.iloc[:, 4], shuffle=True, random_state=42, test_size=0.2)


model = KNeighborsClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print('model accuracy:', accuracy_score(y_test, y_pred) * 100, '%')
```
